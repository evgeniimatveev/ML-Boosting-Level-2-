#⚙️ Overview

**LightGBM (Light Gradient Boosting Machine)** is a high-performance boosting algorithm designed for efficiency, speed, and scalability. It is based on **Gradient Boosting Decision Trees (GBDT)** and is optimized for both large datasets and low-latency applications.

LightGBM is widely used in **competitive ML**, including **Kaggle competitions**, **financial modeling**, **fraud detection**, and **time-series forecasting**.

---

## What is LightGBM?

LightGBM is a decision-tree-based **ensemble learning** method that builds models in a boosting framework. Unlike XGBoost, which grows trees **level-wise**, LightGBM grows trees **leaf-wise**, leading to better accuracy and faster training.

### ⚙️ How it Works:

#### 🌲 Boosting (Gradient Boosting Algorithm):
- Trees are built **sequentially**, each improving the previous model.
- Uses **gradient descent** to minimize the loss function.

#### 🔄 Overfitting Reduction:
- **L1 (Lasso) & L2 (Ridge) regularization** to penalize complex models.
- **Early stopping** prevents overfitting by stopping training when performance stops improving.

#### ⚡ High Performance:
- **Histogram-based splitting** speeds up training.
- **Leaf-wise tree growth** improves accuracy over level-wise growth.
- **Sparse feature optimization** handles missing values efficiently.

#### 🌍 Wide Application:
- Works well for **regression, classification, ranking, and time-series forecasting**.

---

## 📊 Popular Measures in LightGBM:

- **Learning Rate (η) ⚙️** – Controls the step size in gradient descent (lower values prevent overfitting).
- **Max Depth (d) 🌲** – Limits the depth of trees, balancing bias-variance tradeoff.
- **Num Leaves (l) 🍃** – Controls the number of leaves per tree (higher = more complex model).
- **Feature Fraction (f) 🔢** – Controls the fraction of features used per iteration (reduces overfitting).
- **Bagging Fraction (b) 🎒** – Subsampling of data before training (helps generalization).
- **Lambda (λ) & Alpha (α) 🏗️** – L2 & L1 regularization parameters to control model complexity.

---

## ✅ Key Strengths & ❌ Limitations:

### ✅ Key Strengths:
- **High Accuracy:** Performs exceptionally well in Kaggle competitions.
- **Fast Computation:** Uses histogram-based methods for faster training.
- **Handles Large Datasets:** Optimized for distributed computing.
- **Efficient Memory Usage:** Consumes less memory compared to other GBDT models.
- **Built-in Regularization:** L1/L2 penalty reduces overfitting.
- **GPU Support:** Accelerates training on large datasets.

### ❌ Limitations:
- **Sensitive to Hyperparameters:** Needs tuning for best results.
- **Computationally Expensive:** Large datasets require more memory.
- **Not Ideal for Small Data:** Can overfit if dataset is too small.

---

LightGBM is a powerful and efficient model that offers great scalability for structured data tasks! 🚀


## 🔴 LightGBM Formula

LightGBM is based on **Gradient Boosting**, where new trees correct errors of previous trees. The main objective function consists of:

$$
L(\theta) = \sum_{i=1}^{n} l(y_i, \hat{y}_i) + \sum_{k=1}^{K} \Omega(f_k)
$$

where:
$$
 l(y_i, \hat{y}_i) \text{ is the loss function measuring prediction error.} \\
 \Omega(f_k) \text{ is the regularization term controlling model complexity.}
$$

---

### 📉 Loss Function:

LightGBM minimizes a differentiable loss function, such as:

#### 🔹 Mean Squared Error (MSE) for Regression:
$$
L(y, \hat{y}) = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

#### 🔹 Log Loss for Classification:
$$
L(y, \hat{y}) = -\sum_{i=1}^{n} \left[y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i)\right]
$$

---

## 🎛️ Regularization Terms (Shrinkage & Pruning):

The regularization function is:
$$
\Omega(f) = T\gamma + \frac{1}{2} \sum w_j^2
$$

where:
$$
 T \text{ = number of leaves in the tree 🍃} \\
 w_j \text{ = leaf weight ✖️} \\
 \gamma \text{ = min_split_loss (pruning parameter) ❌} \\
 \lambda \text{ = L2 regularization } \lambda \text{ 💰}
$$

---

## 🎯 Choosing the Right Hyperparameters:

| Parameter | Description | Default | Recommended Range 🔧 |
|-----------|-------------|---------|----------------------|
| $$\eta$$ (Learning Rate) | Controls the step size in boosting. Lower = slower but better generalization. | 0.1 | 0.01 – 0.3 |
| $$max\_depth$$ | Maximum depth of trees. Higher = more complex, risk of overfitting. | 6 | 3 – 10 |
| $$subsample$$ | Fraction of samples used per boosting round. Helps with regularization. | 1.0 | 0.5 – 1.0 |
| $$feature\_fraction$$ | Fraction of features per tree. Reduces overfitting. | 1.0 | 0.3 – 1.0 |
| $$\lambda$$ | L2 regularization (Ridge). Controls complexity. Higher = stronger penalty. | 1.0 | 0 – 10 |
| $$\alpha$$ | L1 regularization (Lasso). Helps feature selection. | 0.0 | 0 – 10 |
| $$\gamma$$ | Minimum loss reduction for splits. Higher = conservative splits. | 0.0 | 0 – 5 |

# 📌 Import necessary libraries

```python
# 📌 Import necessary libraries
import numpy as np  # Numerical computations ⚡
import pandas as pd  # Data manipulation 📊
import matplotlib.pyplot as plt  # Visualization 📈
import seaborn as sns  # Advanced plots 🎨
import time  # Time tracking ⏱️
# 📌 Machine learning libraries
import xgboost as xgb  # XGBoost Regressor 🚀
import lightgbm as lgb  # LightGBM Regressor 🌿
from sklearn.ensemble import RandomForestRegressor  # Random Forest 🌲
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score  # Model selection & tuning 🎯
from sklearn.preprocessing import StandardScaler  # Feature scaling 🔄
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score  # Evaluation metrics ✅
```

# 📌 Load dataset from CSV file

```python
# 📌 Load dataset from CSV file
file_path = "insurance.csv"  # Path to the dataset 📂
df = pd.read_csv(file_path)  # Read CSV into a DataFrame 📜
```

```python
df.head()
```

# 📌 Convert categorical features into numerical using One-Hot Encoding

```python
# 📌 Convert categorical features into numerical using One-Hot Encoding
# This converts categorical variables into multiple binary columns 🏷️
df = pd.get_dummies(df, columns=['sex', 'smoker', 'region'], drop_first=True)
```

# 📌 Define feature matrix (X) and target variable (y)

```python
# 📌 Define feature matrix (X) and target variable (y)
X = df.drop(columns=['charges'])  # Feature variables 🛠️
y = df['charges']  # Target variable 🎯
```

# 📌 Standardize the feature matrix to improve model performance

```python
# 📌 Standardize the feature matrix to improve model performance
scaler = StandardScaler()  # Initialize scaler 🔄
X_scaled = scaler.fit_transform(X)  # Scale features 📏
```

# 📌 Split data into 80% training and 20% testing sets

```python
# 📌 Split data into 80% training and 20% testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)  # Train-test split 🧪
```

# 📌📌📌 Initialize three models for comparison

```python
# 📌 Initialize three models for comparison
xgb_regressor = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)  # XGBoost 🚀
lgb_regressor = lgb.LGBMRegressor(random_state=42)  # LightGBM 🌿
rf_regressor = RandomForestRegressor(random_state=42)  # Random Forest 🌲
```

# 📌 Train all models using the training dataset

```python
# 📌 Train all models using the training dataset
xgb_regressor.fit(X_train, y_train)  # Train XGBoost ⚡
lgb_regressor.fit(X_train, y_train)  # Train LightGBM 🌱
rf_regressor.fit(X_train, y_train)  # Train RandomForest 🌳
```

# 📌 Store models in a dictionary for easier iteration

```python
# 📌 Store models in a dictionary for easier iteration
models = {
    "XGBoost": xgb_regressor,
    "LightGBM": lgb_regressor,
    "RandomForest": rf_regressor
}
```

# 📌 List to store evaluation results

```python

# 📌 List to store evaluation results
results = []  # Empty list to store metrics 📊

def evaluate_model(model, model_name):
    """
    📌 Evaluates a given model using multiple performance metrics and appends the results.
    """
    y_train_pred = model.predict(X_train)  # Predictions on training set 🎯
    y_test_pred = model.predict(X_test)  # Predictions on test set 🎯

    # 📌 Compute evaluation metrics for both training and testing sets
    for dataset, y_true, y_pred in zip(["Train", "Test"], [y_train, y_test], [y_train_pred, y_test_pred]):
        metrics = {
            "Model": model_name,
            "Dataset": dataset,
            "MAE": mean_absolute_error(y_true, y_pred),  # Mean Absolute Error 🔢
            "MSE": mean_squared_error(y_true, y_pred),  # Mean Squared Error 🔢
            "RMSE": np.sqrt(mean_squared_error(y_true, y_pred)),  # Root Mean Squared Error 📉
            "R²": r2_score(y_true, y_pred),  # R-Squared Score 📊
            "Adjusted R²": 1 - (1 - r2_score(y_true, y_pred)) * (len(y_true) - 1) / (len(y_true) - X.shape[1] - 1)  # Adjusted R² 🧮
        }
        results.append(metrics)  # Store results in list 📜
```

# 📌 Evaluate all models before hyperparameter tuning

```python
# 📌 Evaluate all models before hyperparameter tuning
for name, model in models.items():
    evaluate_model(model, name)
```

# 🚀 Cross-Validation Evaluation (Before Hyperparameter Tuning)

```python
from sklearn.model_selection import cross_val_score

# 🚀 Cross-Validation Evaluation (Before Hyperparameter Tuning)
cv_results = {}  # Dictionary to store cross-validation scores

for name, model in models.items():
    """
    Perform 5-Fold Cross-Validation for each model.
    This helps evaluate model performance more reliably
    before tuning hyperparameters.
    """
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring="r2")  # R² score with 5-Fold CV
    cv_results[name] = scores.mean()  # Compute mean R² score across folds

# 🔥 Print Cross-Validation Results
print("\n🚀 Cross-Validation Results (Before Hyperparameter Tuning):")
for model, score in cv_results.items():
    print(f"{model}: Mean CV R² = {score:.4f}")
```

# 🔥 Print Cross-Validation Results

```python
# 🔥 Print Cross-Validation Results
print("\n🚀 Cross-Validation Results (Before Hyperparameter Tuning):")
for model, score in cv_results.items():
    print(f"{model}: Mean CV R² = {score:.4f}")
```

# 📌 Convert the results list into a DataFrame for better visualization

```python
# 📌 Convert the results list into a DataFrame for better visualization
metrics_df = pd.DataFrame(results)  # Create DataFrame 📊
print("Performance Before Hyperparameter Tuning:")  # Display results 📢
display(metrics_df)
```

# 📌 Define hyperparameter search space for LightGBM optimization

```python
# 📌 Define hyperparameter search space for LightGBM optimization
param_dist = {
    'num_leaves': [20, 31, 40],  # Control model complexity 🎛️
    'learning_rate': [0.01, 0.05, 0.1],  # Learning rate for gradient descent 📉
    'n_estimators': [100, 200, 500]  # Number of boosting iterations 🔁
}
```

# 📌 Perform hyperparameter tuning using RandomizedSearchCV

```python
# 📌 Perform hyperparameter tuning using RandomizedSearchCV for faster optimization
random_search = RandomizedSearchCV(
    lgb.LGBMRegressor(random_state=42),
    param_distributions=param_dist,
    n_iter=10, cv=5,
    scoring='neg_mean_squared_error',
    random_state=42
)
random_search.fit(X_train, y_train)  # Fit model with different hyperparameter combinations 🏆
```

# 📌 Retrieve best model from tuning

```python
# 📌 Retrieve best model from tuning
best_lgb = random_search.best_estimator_  # Best model 🎯
print("Best Parameters for LightGBM:", random_search.best_params_)  # Display best parameters 🏅
```

# 📌 Evaluate the optimized LightGBM model

```python
# 📌 Evaluate the optimized LightGBM model
evaluate_model(best_lgb, "LightGBM (Tuned)")  # Evaluate best model 🎯
```

# 📌 Convert updated results into DataFrame for comparison

```python
# 📌 Convert updated results into DataFrame for comparison
metrics_df = pd.DataFrame(results)  # Create DataFrame 📊
print("Performance After Hyperparameter Tuning:")  # Display results 📢
display(metrics_df)
```

# ⏳⏳⏳⏳ Training Time Comparison (Simplified)

```python
# Initialize dictionary to store training times ⏳
training_times = {}

# Train models and measure training time ⏱️
for name, model in models.items():
    start_time = time.time()  # Start timer ⏳
    model.fit(X_train, y_train)  # Train the model 🎯
    training_times[name] = time.time() - start_time  # Calculate training duration ⏱️

# Check if training_times has values
print(training_times)  # Debugging step to confirm times are stored

# 📌 Training Time Comparison (Simplified) ⏳
plt.figure(figsize=(8,5))
sns.barplot(x=list(training_times.keys()), y=list(training_times.values()), palette='coolwarm')
plt.xlabel("Model")  # X-axis label 🎯
plt.ylabel("Training Time (seconds)")  # Y-axis label ⏱️
plt.title("Model Training Time Comparison")  # Title 📢

# Display time values above bars for better readability
for index, value in enumerate(training_times.values()):
    plt.text(index, value + 0.1, f"{value:.2f}s", ha='center', fontsize=12, fontweight='bold')

plt.show()
```
