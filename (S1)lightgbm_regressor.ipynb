#‚öôÔ∏è Overview

**LightGBM (Light Gradient Boosting Machine)** is a high-performance boosting algorithm designed for efficiency, speed, and scalability. It is based on **Gradient Boosting Decision Trees (GBDT)** and is optimized for both large datasets and low-latency applications.

LightGBM is widely used in **competitive ML**, including **Kaggle competitions**, **financial modeling**, **fraud detection**, and **time-series forecasting**.

---

## What is LightGBM?

LightGBM is a decision-tree-based **ensemble learning** method that builds models in a boosting framework. Unlike XGBoost, which grows trees **level-wise**, LightGBM grows trees **leaf-wise**, leading to better accuracy and faster training.

### ‚öôÔ∏è How it Works:

#### üå≤ Boosting (Gradient Boosting Algorithm):
- Trees are built **sequentially**, each improving the previous model.
- Uses **gradient descent** to minimize the loss function.

#### üîÑ Overfitting Reduction:
- **L1 (Lasso) & L2 (Ridge) regularization** to penalize complex models.
- **Early stopping** prevents overfitting by stopping training when performance stops improving.

#### ‚ö° High Performance:
- **Histogram-based splitting** speeds up training.
- **Leaf-wise tree growth** improves accuracy over level-wise growth.
- **Sparse feature optimization** handles missing values efficiently.

#### üåç Wide Application:
- Works well for **regression, classification, ranking, and time-series forecasting**.

---

## üìä Popular Measures in LightGBM:

- **Learning Rate (Œ∑) ‚öôÔ∏è** ‚Äì Controls the step size in gradient descent (lower values prevent overfitting).
- **Max Depth (d) üå≤** ‚Äì Limits the depth of trees, balancing bias-variance tradeoff.
- **Num Leaves (l) üçÉ** ‚Äì Controls the number of leaves per tree (higher = more complex model).
- **Feature Fraction (f) üî¢** ‚Äì Controls the fraction of features used per iteration (reduces overfitting).
- **Bagging Fraction (b) üéí** ‚Äì Subsampling of data before training (helps generalization).
- **Lambda (Œª) & Alpha (Œ±) üèóÔ∏è** ‚Äì L2 & L1 regularization parameters to control model complexity.

---

## ‚úÖ Key Strengths & ‚ùå Limitations:

### ‚úÖ Key Strengths:
- **High Accuracy:** Performs exceptionally well in Kaggle competitions.
- **Fast Computation:** Uses histogram-based methods for faster training.
- **Handles Large Datasets:** Optimized for distributed computing.
- **Efficient Memory Usage:** Consumes less memory compared to other GBDT models.
- **Built-in Regularization:** L1/L2 penalty reduces overfitting.
- **GPU Support:** Accelerates training on large datasets.

### ‚ùå Limitations:
- **Sensitive to Hyperparameters:** Needs tuning for best results.
- **Computationally Expensive:** Large datasets require more memory.
- **Not Ideal for Small Data:** Can overfit if dataset is too small.

---

LightGBM is a powerful and efficient model that offers great scalability for structured data tasks! üöÄ


## üî¥ LightGBM Formula

LightGBM is based on **Gradient Boosting**, where new trees correct errors of previous trees. The main objective function consists of:

$$
L(\theta) = \sum_{i=1}^{n} l(y_i, \hat{y}_i) + \sum_{k=1}^{K} \Omega(f_k)
$$

where:
$$
 l(y_i, \hat{y}_i) \text{ is the loss function measuring prediction error.} \\
 \Omega(f_k) \text{ is the regularization term controlling model complexity.}
$$

---

### üìâ Loss Function:

LightGBM minimizes a differentiable loss function, such as:

#### üîπ Mean Squared Error (MSE) for Regression:
$$
L(y, \hat{y}) = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

#### üîπ Log Loss for Classification:
$$
L(y, \hat{y}) = -\sum_{i=1}^{n} \left[y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i)\right]
$$

---

## üéõÔ∏è Regularization Terms (Shrinkage & Pruning):

The regularization function is:
$$
\Omega(f) = T\gamma + \frac{1}{2} \sum w_j^2
$$

where:
$$
 T \text{ = number of leaves in the tree üçÉ} \\
 w_j \text{ = leaf weight ‚úñÔ∏è} \\
 \gamma \text{ = min_split_loss (pruning parameter) ‚ùå} \\
 \lambda \text{ = L2 regularization } \lambda \text{ üí∞}
$$

---

## üéØ Choosing the Right Hyperparameters:

| Parameter | Description | Default | Recommended Range üîß |
|-----------|-------------|---------|----------------------|
| $$\eta$$ (Learning Rate) | Controls the step size in boosting. Lower = slower but better generalization. | 0.1 | 0.01 ‚Äì 0.3 |
| $$max\_depth$$ | Maximum depth of trees. Higher = more complex, risk of overfitting. | 6 | 3 ‚Äì 10 |
| $$subsample$$ | Fraction of samples used per boosting round. Helps with regularization. | 1.0 | 0.5 ‚Äì 1.0 |
| $$feature\_fraction$$ | Fraction of features per tree. Reduces overfitting. | 1.0 | 0.3 ‚Äì 1.0 |
| $$\lambda$$ | L2 regularization (Ridge). Controls complexity. Higher = stronger penalty. | 1.0 | 0 ‚Äì 10 |
| $$\alpha$$ | L1 regularization (Lasso). Helps feature selection. | 0.0 | 0 ‚Äì 10 |
| $$\gamma$$ | Minimum loss reduction for splits. Higher = conservative splits. | 0.0 | 0 ‚Äì 5 |

# üìå Import necessary libraries

```python
# üìå Import necessary libraries
import numpy as np  # Numerical computations ‚ö°
import pandas as pd  # Data manipulation üìä
import matplotlib.pyplot as plt  # Visualization üìà
import seaborn as sns  # Advanced plots üé®
import time  # Time tracking ‚è±Ô∏è
# üìå Machine learning libraries
import xgboost as xgb  # XGBoost Regressor üöÄ
import lightgbm as lgb  # LightGBM Regressor üåø
from sklearn.ensemble import RandomForestRegressor  # Random Forest üå≤
from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score  # Model selection & tuning üéØ
from sklearn.preprocessing import StandardScaler  # Feature scaling üîÑ
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score  # Evaluation metrics ‚úÖ
```

# üìå Load dataset from CSV file

```python
# üìå Load dataset from CSV file
file_path = "insurance.csv"  # Path to the dataset üìÇ
df = pd.read_csv(file_path)  # Read CSV into a DataFrame üìú
```

```python
df.head()
```

# üìå Convert categorical features into numerical using One-Hot Encoding

```python
# üìå Convert categorical features into numerical using One-Hot Encoding
# This converts categorical variables into multiple binary columns üè∑Ô∏è
df = pd.get_dummies(df, columns=['sex', 'smoker', 'region'], drop_first=True)
```

# üìå Define feature matrix (X) and target variable (y)

```python
# üìå Define feature matrix (X) and target variable (y)
X = df.drop(columns=['charges'])  # Feature variables üõ†Ô∏è
y = df['charges']  # Target variable üéØ
```

# üìå Standardize the feature matrix to improve model performance

```python
# üìå Standardize the feature matrix to improve model performance
scaler = StandardScaler()  # Initialize scaler üîÑ
X_scaled = scaler.fit_transform(X)  # Scale features üìè
```

# üìå Split data into 80% training and 20% testing sets

```python
# üìå Split data into 80% training and 20% testing sets
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)  # Train-test split üß™
```

# üìåüìåüìå Initialize three models for comparison

```python
# üìå Initialize three models for comparison
xgb_regressor = xgb.XGBRegressor(objective='reg:squarederror', random_state=42)  # XGBoost üöÄ
lgb_regressor = lgb.LGBMRegressor(random_state=42)  # LightGBM üåø
rf_regressor = RandomForestRegressor(random_state=42)  # Random Forest üå≤
```

# üìå Train all models using the training dataset

```python
# üìå Train all models using the training dataset
xgb_regressor.fit(X_train, y_train)  # Train XGBoost ‚ö°
lgb_regressor.fit(X_train, y_train)  # Train LightGBM üå±
rf_regressor.fit(X_train, y_train)  # Train RandomForest üå≥
```

# üìå Store models in a dictionary for easier iteration

```python
# üìå Store models in a dictionary for easier iteration
models = {
    "XGBoost": xgb_regressor,
    "LightGBM": lgb_regressor,
    "RandomForest": rf_regressor
}
```

# üìå List to store evaluation results

```python

# üìå List to store evaluation results
results = []  # Empty list to store metrics üìä

def evaluate_model(model, model_name):
    """
    üìå Evaluates a given model using multiple performance metrics and appends the results.
    """
    y_train_pred = model.predict(X_train)  # Predictions on training set üéØ
    y_test_pred = model.predict(X_test)  # Predictions on test set üéØ

    # üìå Compute evaluation metrics for both training and testing sets
    for dataset, y_true, y_pred in zip(["Train", "Test"], [y_train, y_test], [y_train_pred, y_test_pred]):
        metrics = {
            "Model": model_name,
            "Dataset": dataset,
            "MAE": mean_absolute_error(y_true, y_pred),  # Mean Absolute Error üî¢
            "MSE": mean_squared_error(y_true, y_pred),  # Mean Squared Error üî¢
            "RMSE": np.sqrt(mean_squared_error(y_true, y_pred)),  # Root Mean Squared Error üìâ
            "R¬≤": r2_score(y_true, y_pred),  # R-Squared Score üìä
            "Adjusted R¬≤": 1 - (1 - r2_score(y_true, y_pred)) * (len(y_true) - 1) / (len(y_true) - X.shape[1] - 1)  # Adjusted R¬≤ üßÆ
        }
        results.append(metrics)  # Store results in list üìú
```

# üìå Evaluate all models before hyperparameter tuning

```python
# üìå Evaluate all models before hyperparameter tuning
for name, model in models.items():
    evaluate_model(model, name)
```

# üöÄ Cross-Validation Evaluation (Before Hyperparameter Tuning)

```python
from sklearn.model_selection import cross_val_score

# üöÄ Cross-Validation Evaluation (Before Hyperparameter Tuning)
cv_results = {}  # Dictionary to store cross-validation scores

for name, model in models.items():
    """
    Perform 5-Fold Cross-Validation for each model.
    This helps evaluate model performance more reliably
    before tuning hyperparameters.
    """
    scores = cross_val_score(model, X_train, y_train, cv=5, scoring="r2")  # R¬≤ score with 5-Fold CV
    cv_results[name] = scores.mean()  # Compute mean R¬≤ score across folds

# üî• Print Cross-Validation Results
print("\nüöÄ Cross-Validation Results (Before Hyperparameter Tuning):")
for model, score in cv_results.items():
    print(f"{model}: Mean CV R¬≤ = {score:.4f}")
```

# üî• Print Cross-Validation Results

```python
# üî• Print Cross-Validation Results
print("\nüöÄ Cross-Validation Results (Before Hyperparameter Tuning):")
for model, score in cv_results.items():
    print(f"{model}: Mean CV R¬≤ = {score:.4f}")
```

# üìå Convert the results list into a DataFrame for better visualization

```python
# üìå Convert the results list into a DataFrame for better visualization
metrics_df = pd.DataFrame(results)  # Create DataFrame üìä
print("Performance Before Hyperparameter Tuning:")  # Display results üì¢
display(metrics_df)
```

# üìå Define hyperparameter search space for LightGBM optimization

```python
# üìå Define hyperparameter search space for LightGBM optimization
param_dist = {
    'num_leaves': [20, 31, 40],  # Control model complexity üéõÔ∏è
    'learning_rate': [0.01, 0.05, 0.1],  # Learning rate for gradient descent üìâ
    'n_estimators': [100, 200, 500]  # Number of boosting iterations üîÅ
}
```

# üìå Perform hyperparameter tuning using RandomizedSearchCV

```python
# üìå Perform hyperparameter tuning using RandomizedSearchCV for faster optimization
random_search = RandomizedSearchCV(
    lgb.LGBMRegressor(random_state=42),
    param_distributions=param_dist,
    n_iter=10, cv=5,
    scoring='neg_mean_squared_error',
    random_state=42
)
random_search.fit(X_train, y_train)  # Fit model with different hyperparameter combinations üèÜ
```

# üìå Retrieve best model from tuning

```python
# üìå Retrieve best model from tuning
best_lgb = random_search.best_estimator_  # Best model üéØ
print("Best Parameters for LightGBM:", random_search.best_params_)  # Display best parameters üèÖ
```

# üìå Evaluate the optimized LightGBM model

```python
# üìå Evaluate the optimized LightGBM model
evaluate_model(best_lgb, "LightGBM (Tuned)")  # Evaluate best model üéØ
```

# üìå Convert updated results into DataFrame for comparison

```python
# üìå Convert updated results into DataFrame for comparison
metrics_df = pd.DataFrame(results)  # Create DataFrame üìä
print("Performance After Hyperparameter Tuning:")  # Display results üì¢
display(metrics_df)
```
