# üîç Overview

**XGBoost Classifier** is a powerful **ensemble learning algorithm** based on **Gradient Boosting Decision Trees (GBDT)**.  
It is designed for **high performance, accuracy, and scalability**, making it one of the best models for **classification tasks**.

XGBoost is widely used in **fraud detection, medical diagnosis, image classification, and sentiment analysis**.

---

## üî• What is XGBoost?

XGBoost is a **tree-based ensemble learning method** that builds models in a **boosting** framework.  
**Unlike Random Forest, which builds trees independently**, XGBoost **builds trees sequentially**, learning from previous errors to improve accuracy.

---

## ‚öôÔ∏è How It Works?

### üöÄ Boosting (Gradient Boosting Algorithm)
‚úîÔ∏è Trees are built **sequentially**, each improving the previous model.  
‚úîÔ∏è Uses **gradient descent** to minimize the loss function.

### üéØ Overfitting Reduction
‚úîÔ∏è **L1 (Lasso) & L2 (Ridge) regularization** to penalize complex models.  
‚úîÔ∏è **Early stopping** prevents overfitting by stopping training when performance stops improving.

### üí° High Performance
‚úîÔ∏è **Parallel computation** for faster training.  
‚úîÔ∏è **Sparse-aware algorithms** handle missing values automatically.  
‚úîÔ∏è **Tree pruning** improves efficiency by removing unimportant splits.

### üîÑ Applications
XGBoost **performs exceptionally well in classification tasks**, including:  
‚úîÔ∏è **Fraud detection (Finance & Banking)**  
‚úîÔ∏è **Medical diagnosis (Healthcare & Biotech)**  
‚úîÔ∏è **Spam filtering (Email & Messaging Services)**  
‚úîÔ∏è **Sentiment analysis (Social Media & NLP)**  
‚úîÔ∏è **Image classification (Computer Vision)**  

---

## üìä Key Hyperparameters for XGBoost Classifier:

‚úîÔ∏è **Learning Rate (Œ∑)** üîΩ ‚Äî Controls the step size in gradient descent (**lower values prevent overfitting**).  
‚úîÔ∏è **Max Depth (D)** üå≤ ‚Äî Limits the depth of trees, balancing bias-variance tradeoff.  
‚úîÔ∏è **Subsample (S)** üéØ ‚Äî Controls the fraction of data used for training each tree (**reduces overfitting**).  
‚úîÔ∏è **Lambda (Œª) & Alpha (Œ±)** ‚ûñ ‚Äî L2 & L1 regularization parameters to control model complexity.  
‚úîÔ∏è **Gamma (Œ≥)** ‚úÇÔ∏è ‚Äî Minimum loss reduction required to make a split (**higher = more conservative splits**).  
‚úîÔ∏è **Objective Function** üéØ ‚Äî Defines the classification task (e.g., `"binary:logistic"` for binary classification, `"multi:softmax"` for multi-class).  

---

## ‚úÖ Key Strengths & Limitations:

### üèÜ Strengths:
‚úîÔ∏è **High accuracy** ‚Äî Used in top Kaggle classification solutions.  
‚úîÔ∏è **Fast computation** ‚Äî Parallelization and optimized memory usage.

# ‚ö° XGBoost Formula

XGBoost is based on **Gradient Boosting**, where new trees correct errors of previous trees.  
The main objective function consists of:

$$
L(\theta) = \sum_{i=1}^{n} L(y_i, \hat{y}_i) + \sum_{k=1}^{K} \Omega(f_k)
$$

Where:  
- \( L(y_i, \hat{y}_i) \) is the **loss function** measuring prediction error.  
- \( \Omega(f_k) \) is the **regularization term** controlling model complexity.  

---

## üéØ Loss Function:

XGBoost minimizes a **differentiable loss function**, such as:

- **Log Loss for classification**:

$$
L(y, \hat{y}) = -\sum_{i=1}^{n} \left[y_i \log(\hat{y}_i) + (1 - y_i) \log(1 - \hat{y}_i) \right]
$$

Used in **binary classification** tasks (e.g., fraud detection, spam filtering).

- **Multiclass Log Loss (Softmax Loss) for multi-class classification**:

$$
L(y, \hat{y}) = -\sum_{i=1}^{n} \sum_{k=1}^{K} y_{ik} \log(\hat{y}_{ik})
$$

Used for **multi-class classification problems** (e.g., image recognition, NLP tasks).

---

## üèóÔ∏è Regularization Terms (Shrinkage & Pruning):

The regularization function in XGBoost:

$$
\Omega(f) = T\gamma + \frac{1}{2} \lambda \sum w_j^2
$$

Where:  
- \( T \) = Number of leaves in the tree üåø  
- \( w_j \) = Leaf weight üèãÔ∏è‚Äç‚ôÇÔ∏è  
- \( \gamma \) = `min_split_loss` (Pruning parameter) ‚ùå  
- \( \lambda \) = L2 regularization (Ridge penalty) üõ°  

---

## üéõ Choosing the Right Hyperparameters:

| Parameter        | Description | Default | Recommended Range |
|-----------------|-------------|---------|------------------|
| **\( \eta \) (Learning Rate)** | Controls step size in boosting. Lower = slower but better generalization. | `0.3` | `0.01 ‚Äì 0.3` |
| **max_depth** | Maximum depth of trees. Higher = more complex model, risk of overfitting. | `6` | `3 ‚Äì 10` |
| **subsample** | Fraction of features per tree. Reduces overfitting. | `1.0` | `0.5 ‚Äì 1.0` |
| **colsample_bytree** | Fraction of features per split. Reduces overfitting. | `1.0` | `0.3 ‚Äì 1.0` |
| **\( \lambda \) (L2 regularization)** | L2 regularization (Ridge). Controls complexity. Higher = stronger penalty. | `1.0` | `0 ‚Äì 10` |
| **\( \alpha \) (L1 regularization)** | L1 regularization (Lasso). Helps in feature selection. | `0.0` | `0 ‚Äì 10` |
| **\( \gamma \) (Gamma)** | Minimum loss reduction for splits. Higher = more conservative splits. | `0.0` | `0 ‚Äì 5` |

## Part 1 - Data Preprocessing

```
# üìå Importing necessary libraries
import numpy as np  # NumPy for numerical computations
import pandas as pd  # Pandas for data manipulation
import matplotlib.pyplot as plt  # Matplotlib for data visualization
import seaborn as sns  # Seaborn for advanced visualization

# Importing Scikit-learn modules
from sklearn.model_selection import train_test_split, cross_val_score, GridSearchCV  # Model selection and evaluation
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder  # Data preprocessing
from sklearn.compose import ColumnTransformer  # Column transformation
from sklearn.pipeline import Pipeline  # Pipeline for streamlining preprocessing and modeling
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, roc_auc_score  # Model performance metrics

import xgboost as xgb  # XGBoost library for optimized gradient boosting
from IPython.display import display  # Display function for better visualization in Jupyter Notebook
```

# üéØ Load Dataset

```
# üéØ Load Dataset
df = pd.read_csv("churn_modelling.csv")
```

# üìå Drop unnecessary columns (Surname not needed)

```
# üìå Drop unnecessary columns (Surname not needed)
df.drop(columns=['Surname'], inplace=True)
```

# üìå Encoding categorical variables

```
# üìå Encoding categorical variables
# Label Encoding for 'Gender'
label_encoder = LabelEncoder()
df['Gender'] = label_encoder.fit_transform(df['Gender'])  # Male = 1, Female = 0

# One-Hot Encoding for 'Geography'
df = pd.get_dummies(df, columns=['Geography'], drop_first=True)  # Avoid dummy variable trap
```

# üìå Feature Selection (Last column is target)

```
# üìå Feature Selection (Last column is target)
X = df.drop(columns=['Exited']).values   # Features (everything except the target)
y = df['Exited'].values  # Target variable
```

# üìå Feature Scaling

```
# üìå Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

# üìå Train-Test Split (80-20 ratio)

```
# üìå Train-Test Split (80-20 ratio)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42, stratify=y)
```

# üìå Applying K-Fold Cross-Validation (Default K=5)

```
# üìå Applying K-Fold Cross-Validation (Default K=5)

# Initialize the XGBoost classifier with binary logistic objective and log loss as the evaluation metric
model = xgb.XGBClassifier(objective="binary:logistic", eval_metric="logloss")

# Perform 5-fold cross-validation using accuracy as the scoring metric
cv_scores = cross_val_score(model, X_train, y_train, cv=5, scoring='accuracy')

# Print the mean accuracy and standard deviation across folds
print(f"üîÑ Mean CV Accuracy: {cv_scores.mean():.4f} ¬± {cv_scores.std():.4f}")
```

# üìå Train Initial XGBoost Model

```
# üìå Train Initial XGBoost Model
model.fit(X_train, y_train)
```

# üìä Model Evaluation (Before Tuning)

```
# üìä Model Evaluation (Before Tuning)
y_train_pred = model.predict(X_train)
y_test_pred = model.predict(X_test)

train_acc = accuracy_score(y_train, y_train_pred)
test_acc = accuracy_score(y_test, y_test_pred)

print(f"‚úÖ Train Accuracy: {train_acc:.4f}")
print(f"‚úÖ Test Accuracy: {test_acc:.4f}")
```

# üéØ Confusion Matrix

```
# üéØ Confusion Matrix
train_cm = confusion_matrix(y_train, y_train_pred)
test_cm = confusion_matrix(y_test, y_test_pred)
```

# üìä Plot Confusion Matrices

```
# üìä Plot Confusion Matrices
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

sns.heatmap(train_cm, annot=True, fmt="d", cmap="Blues", ax=axes[0])
axes[0].set_title("Train Confusion Matrix")

sns.heatmap(test_cm, annot=True, fmt="d", cmap="Oranges", ax=axes[1])
axes[1].set_title("Test Confusion Matrix")

plt.show()
```

# üìå Classification Report

```
# üìå Classification Report
print("\nüîç Classification Report (Test Set):\n", classification_report(y_test, y_test_pred))
```

# üìå ROC-AUC Score

```
# üìå ROC-AUC Score
roc_auc = roc_auc_score(y_test, y_test_pred)
print(f"üöÄ ROC-AUC Score: {roc_auc:.4f}")
```

# üéØ Hyperparameter Tuning (GridSearchCV)

# üìåüéØüéØ Define hyperparameter grid for tuning

```
# üìå Define hyperparameter grid for tuning
# Specifying the hyperparameters and their possible values for tuning
param_grid = {
    'max_depth': [3, 4, 5],  # Maximum depth of trees (controls complexity and overfitting)
    'learning_rate': [0.01, 0.05, 0.1],  # Step size shrinkage to prevent overfitting
    'n_estimators': [30, 50, 100]  # Number of boosting rounds (trees)
}

# üìå Perform Grid Search CV
# Grid search to find the best combination of hyperparameters using cross-validation
grid_search = GridSearchCV(
    estimator=xgb.XGBClassifier(objective="binary:logistic", eval_metric="logloss"),  # XGBoost classifier
    param_grid=param_grid,  # Hyperparameter grid
    scoring='accuracy',  # Evaluation metric (maximize accuracy)
    cv=5,  # 5-fold cross-validation
    verbose=1  # Display progress during training
)

# Fit grid search to training data
grid_search.fit(X_train, y_train)  # Training the model with all parameter combinations

# üìå Print best hyperparameters after tuning
print(f"üéØ Best Hyperparameters: {grid_search.best_params_}")
```

```
# üéØ Hyperparameter Tuning (GridSearchCV)
best_model = xgb.XGBClassifier(
    max_depth=4,                # Reduce tree depth to prevent overfitting
    learning_rate=0.05,         # Slow learning rate for better generalization
    n_estimators=30,            # Reduce the number of trees
    subsample=0.9,              # Use 90% of data per tree
    colsample_bytree=0.8,       # Use 80% of features per split
    scale_pos_weight=2,         # Balance class weights (can be calculated)
    early_stopping_rounds=10,   # Stop early if no improvement
    verbosity=0,
    objective="binary:logistic",
    eval_metric="logloss"
)

best_model.fit(X_train, y_train, eval_set=[(X_test, y_test)], verbose=False)
```

# ‚úÖ Best Hyperparameters

```
# ‚úÖ Best Hyperparameters
print(f"üéØ Best Hyperparameters: {grid_search.best_params_}")
```

# üìå Train Model with Best Hyperparameters

```
# üìå Train Model with Best Hyperparameters
best_model = xgb.XGBClassifier(**grid_search.best_params_, objective="binary:logistic", eval_metric="logloss")
best_model.fit(X_train, y_train)
```

# üìä Evaluate Tuned Model

```
# üìä Evaluate Tuned Model
y_train_pred_tuned = best_model.predict(X_train)
y_test_pred_tuned = best_model.predict(X_test)

train_acc_tuned = accuracy_score(y_train, y_train_pred_tuned)
test_acc_tuned = accuracy_score(y_test, y_test_pred_tuned)
```

# üìå Comparison Table

```
# üìå Comparison Table
comparison_df = pd.DataFrame({
    "Metric": ["Accuracy", "Precision", "Recall", "F1-Score", "ROC-AUC"],
    "Before Tuning": [
        accuracy_score(y_test, y_test_pred),
        classification_report(y_test, y_test_pred, output_dict=True)["1"]["precision"],
        classification_report(y_test, y_test_pred, output_dict=True)["1"]["recall"],
        classification_report(y_test, y_test_pred, output_dict=True)["1"]["f1-score"],
        roc_auc
    ],
    "After Tuning": [
        accuracy_score(y_test, y_test_pred_tuned),
        classification_report(y_test, y_test_pred_tuned, output_dict=True)["1"]["precision"],
        classification_report(y_test, y_test_pred_tuned, output_dict=True)["1"]["recall"],
        classification_report(y_test, y_test_pred_tuned, output_dict=True)["1"]["f1-score"],
        roc_auc_score(y_test, y_test_pred_tuned)
    ]
})
```
